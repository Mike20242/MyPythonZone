# Exercise 4.1: NumPy Operations
In this exercise, we will explore basic NumPy operations including creating arrays and performing simple calculations.

```python
import numpy as np

# 1. Create a 1D array from a Python list
scores_1d = np.array([85, 90, 78, 92, 88])
print("1D Array of Scores:", scores_1d)
print("Shape:", scores_1d.shape)
```

```python
# 2. Create a 2D array (e.g., representing scores of 3 students in 2 subjects)
scores_2d = np.array([
    [85, 90],
    [78, 85],
    [92, 95]
])
print("\n2D Array of Scores (Rows: Students, Columns: Subjects):\n", scores_2d)
print("Shape:", scores_2d.shape)
```

```python
# 3. Basic Calculations on 1D Array
print("\n--- Basic Calculations (1D Array) ---")
print("Total sum of scores:", np.sum(scores_1d))
print("Average score:", np.mean(scores_1d))
print("Highest score:", np.max(scores_1d))
print("Lowest score:", np.min(scores_1d))
```

```python
# 4. Calculations on 2D Array (along axes)
print("\n--- Axis-based Calculations (2D Array) ---")
print("Average score per student (across subjects):", np.mean(scores_2d, axis=1))
print("Average score per subject (across students):", np.mean(scores_2d, axis=0))
```

```python
# 5. Element-wise operations (e.g., adding 5 bonus points to all scores)
print("\n--- Element-wise Operations ---")
bonus_scores = scores_1d + 5
print("Original scores:", scores_1d)
print("Scores with +5 bonus:", bonus_scores)
```
