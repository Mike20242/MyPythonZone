# Exercise 4.2: Pandas Introduction
In this exercise, we will load a CSV file into a Pandas DataFrame and explore its structure.

```python
import pandas as pd

# 1. Load data from CSV file
# Ensure 'student_data.csv' is in the same folder as this notebook
df = pd.read_csv('student_data.csv')
print("Data loaded successfully!")
```

```python
# 2. View the first few rows of the DataFrame
print("First 5 rows of the dataset:")
display(df.head())
```

```python
# 3. Explore the data structure and data types
print("\nData Information (Columns, Non-Null Count, Dtypes):")
df.info()
```

```python
# 4. Generate summary statistics for numerical columns
print("\nSummary Statistics:")
display(df.describe())
```

```python
# 5. Check the shape of the DataFrame (rows, columns)
print("\nShape of the DataFrame:", df.shape)
print(f"There are {df.shape[0]} rows and {df.shape[1]} columns.")
```
